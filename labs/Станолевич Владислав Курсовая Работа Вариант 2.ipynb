```python
#ะะฐะดะฐัะฐ 5

def goodStr(s, n, c):                         #ะพะฟัะตะดะตะปัะตั ะผะธะฝะธะผะฐะปัะฝะพะต ะบะพะป-ะฒะพ ะธะทะผะตะฝะตะฝะธะน, ััะตะฑัะตะผะพะต ะดะปั ัะพะทะดะฐะฝะธั ัะพัะพัะตะน ัััะพะบะธ
    if n == 1:
        if s[0] != c:                          
            return 1                          #1-e ััะปะพะฒะธะต
        else:
            return 0                          
    
    mid = n // 2
    
    s1 = s[:n // 2]                           #ะฟัะพะฒะตััะตะผ ะฟะตัะฒัั ะฟะพะปะพะฒะธะฝั
    l1=(s1.count(c))                          #ะบะพะป-ะฒะพ ะฒัะพะถะดะตะฝะธะน ัะธะผะฒะพะปะฐ c ะฒ s
    
    s2 = s[n // 2:]                           #ะฟัะพะฒะตััะตะผ ะฒัะพััั ะฟะพะปะพะฒะธะฝั                     
    r2 = (s2.count(c))
    
    secondCon = mid - l1 + goodStr(s2,n // 2,chr(ord(c) + 1))          #2-ะต ััะปะพะฒะธะต
    thirdCon = mid - r2 + goodStr(s1,n // 2,chr(ord(c) + 1))          #3-ะต ััะปะพะฒะธะต
    return(min(secondCon,thirdCon))
    
t = int(input())

for i in range(t):
    n = int(input())
    s = input()
    print(goodStr(s, n, 'a'))
    
```
