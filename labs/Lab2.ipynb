```python
#Бинарный поиск
#Сложность - log2N, N - кол-во элементов массива

import random
import time

def divide(arr:list, needle:int, start:int, end:int):
    if (end < start):
        return -1
    
    mid = (start + end) >> 1
    
    if (arr[mid] == needle):
        return mid
    elif (arr[mid] > needle):
        return divide(arr, needle, start, mid - 1)
    else:
        return divide(arr, needle, mid + 1, end)
    
def find(arr : list, needle : int):
    sortedArr = sorted(arr)
    return divide(sortedArr, needle, 0, len(arr)-1)


arr = [11, 44, 12, 8, 73, 111, 2, 586, 98, 3, 5, 400]
arr.sort()
print(arr, " - исходный массив")
needle = 111

start_time = (time.perf_counter())
myRes = find(arr, needle)
end_time = (time.perf_counter())

print("Исходное значение:", needle)
print("Результат бинарного поиска: ", myRes)
print("Время бинарного поиска: {:.6f}ms".format((end_time-start_time)*100))

print("===========================")

start_time = (time.perf_counter())
stdRes = arr.index(needle)
end_time = (time.perf_counter())

print("Результат стандартного поиска: ", stdRes)
print("Время стандартного поиска: {:.6f}ms".format((end_time-start_time)*100))
```

```python
#Бинарное дерево

class Tree:
    right = None
    left = None
    
    def __init__(self, root:int, index:int, parent = None):
        self.root = root
        self.index = index
        self.parent = parent
        self.left = None
        self.right = None
        
    def add(self, value: int, index:int):     
        if value < self.root:
            if self.left == None:
                self.left = Tree(value, index, self)
            else:
                self.left.add(value, index)
        else:
            if self.right == None:
                self.right = Tree(value, index, self)
            else:
                self.right.add(value, index)
    
    def printTree(self):
        if self.left != None:
            self.left.printTree()
        if self.parent != None:
            print(self.root, "=>", self.parent.root)
        else:
            print(self.root, "<= корень")
        if self.right != None:
            self.right.printTree()
    
    def find(self, value:int):
        if self.root == value:
            return self.index
        elif value < self.root:
            if self.left == None:
                return -1
            else:
                return self.left.find(value)
        else:
            if self.right == None:
                return -1
            else:
                return self.right.find(value)
    def delete(self, value):
        if self.root == value:
            if self.right == None and self.left == None:
                newNode = None
            elif self.right == None:
                newNode = self.left
            elif self.left == None:
                newNode = self.right
            else:
                newNode = self.right.findMinEl()
                
            if self.parent != None:     
                if self.parent.right == self:
                    if newNode != None:
                        self.parent.right.root = newNode.root
                    else:
                        self.parent.right = None
                else:
                    if newNode != None:
                        self.parent.left.root = newNode.root
                    else:
                        self.parent.left = None
            elif newNode != None:
                self.root = newNode.root
            
            if newNode != None:
                newNode.delete(newNode.root)
            
        elif value < self.root:
            if self.left == None:
                return -1
            else:
                self.left.delete(value)
        else:
            if self.right == None:
                return -1
            else:
                self.right.delete(value)
                
    def findMinEl(self):
        if self.left == None:
            return self
        else:
            return self.left.findMinEl()
    
    def __cmp__(self, other):
        if self.root == other.root:
            return 0
        elif self.root < other.root:
            return -1
        else:
            return 1
        

tree = Tree(arr[0], 0)

for i in range(1, len(arr)):
    tree.add(arr[i], i)

print(arr, " - исходный массив")

start_time = (time.perf_counter())
myRes = tree.find(needle)
end_time = (time.perf_counter())

print("Исходное значение:", needle)
print("Результат поиска по бинарному дереву: ", myRes)
print("Время поиска по бинарному дереву: {:.6f}ms".format((end_time-start_time)*100))

print("===========================")

start_time = (time.perf_counter())
stdRes = arr.index(needle)
end_time = (time.perf_counter())

print("Результат стандартного поиска: ", stdRes)
print("Время стандартного поиска: {:.6f}ms".format((end_time-start_time)*100))
```

```python
#Метод Фибоначчи
def fibSeq(k:int):
    if k in [0, 1]:
        return k
    return fibSeq(k-1) + fibSeq(k-2)

def fibonacciFind(arr:list, needle:int):
    fNum = 0
    k = 0
    n = len(arr)
    
    while(fNum < n + 1):
        k += 1
        fNum = fibSeq(k+1)
    
    m = fNum - n - 1
    i = fibSeq(k) - m
    p = fibSeq(k-1)
    q = fibSeq(k-2)
   
    while(True):
        if (i < 0) or (i in range(n) and arr[i] < needle):
            if p == 1:
                return -1
            else:
                i += q
                p = p - q
                q = q - p
                continue
        elif (i >= n) or (i in range(n) and arr[i] > needle):
            if q == 0:
                return -1
            else:
                i -= q
                p, q = q, p-q
                continue
        else:
            if arr[i] == needle:
                return i
            
print(arr, " - исходный массив")

start_time = (time.perf_counter())
myRes = fibonacciFind(arr, needle)
end_time = (time.perf_counter())

print("Исходное значение:", needle)
print("Результат поиска фибоначчи: ", myRes)
print("Время поиска фибоначчи: {:.6f}ms".format((end_time-start_time)*100))

print("===========================")

start_time = (time.perf_counter())
stdRes = arr.index(needle)
end_time = (time.perf_counter())

print("Результат стандартного поиска: ", stdRes)
print("Время стандартного поиска: {:.6f}ms".format((end_time-start_time)*100))
```

```python
#Интерполяционный
def interpolatingFind(arr:list, needle:int):
    left = 0
    right = len(arr) - 1
    
    while(arr[left] < needle and arr[right] > needle):
        if arr[right] == arr[left]:
            break
        mid = int(left + ((needle - arr[left]) * (right - left)) / (arr[right] - arr[left]))
        
        if arr[mid] > needle:
            right = mid - 1
        elif arr[mid] < needle:
            left = mid + 1
        else:
            return mid
    if arr[right] == needle:
        return right
    elif arr[left] == needle:
        return left
    else:
        return -1
    
print(arr, " - исходный массив")

start_time = (time.perf_counter())
myRes = interpolatingFind(arr, needle)
end_time = (time.perf_counter())

print("Исходное значение:", needle)
print("Результат интерполяционного поиска: ", myRes)
print("Время интерполяционного поиска: {:.6f}ms".format((end_time-start_time)*100))

print("===========================")

start_time = (time.perf_counter())
stdRes = arr.index(needle)
end_time = (time.perf_counter())

print("Результат стандартного поиска: ", stdRes)
print("Время стандартного поиска: {:.6f}ms".format((end_time-start_time)*100))
```

```python
#Рехэширование с помощью псевдослучайных чисел
import random

class HashClass:
    keys = []
    hashes = []
    
    def __init__(self, arr : list):
        self.appendArray(arr)
        
    def appendKey(self, item, length = None):
        if (length == None):
            length = len(self.keys)
            
        self._checkKeyExists(item)
        temp = hash(item) % length
        
        if self._checkCollisions(temp):
            temp = self._pseudoRandRehash(item, length)
            
        self.hashes.append(temp)
        self.keys.append(item)
        
    def printHashes(self):
        for i in range(len(self.keys)):
            print(self.hashes[i], self.keys[i])
        print()
        
    def appendArray(self, arr:list):
        for item in arr:
            self.appendKey(item, len(arr))
            
    def _pseudoRandRehash(self, key, length : int):
        possibleValues = [i for i in range(length * 2)]
        random.shuffle(possibleValues)
        while True:
            i = possibleValues.pop()
            temp = hash(key) % length + i
            if not self._checkCollisions(temp):
                return temp
            
    def _checkCollisions(self, hashValue):
        return hashValue in self.hashes

    def _checkKeyExists(self, key):
        if key in self.keys:
            raise Exception("Key '{}' already exists".format(key))
            
values = ['superstr', 'megastr', 'duperstr', 'supermegastr', 'supermegaduperstr', 'megaduperstr', 'test']
hashes = HashClass(arr=values)

try:
    hashes.printHashes()
    hashes.appendKey('newKey')
    hashes.appendKey('newKey2')
    hashes.appendKey('newKey3')
    hashes.appendKey('newKey4')
    hashes.printHashes()
    
    # тут будет ошибка, т.к ключ уже есть
    hashes.appendKey('newKey')
except Exception as inst:
    print(inst.args)
```

```python
#Метод цепочек
class ChainsMethod:
    _hashes = {}
    _values = {}
    _pointer = 0
    _N = 10
    
    def __init__(self, arr:dict = None):
        if arr != None:
            self.appendArray(arr)
    
    def append(self, key, value):
        keyHash = hash(key) % self._N
        if self._checkCollisions(keyHash):
            prevInd = self._hashes[keyHash]
            self._values[self._pointer] = [key, value, prevInd]
        else:
            self._values[self._pointer] = [key, value]
            
        self._hashes[keyHash] = self._pointer
        self._pointer += 1
            
    def appendArray(self, arr:dict):
        for key, value in arr.items():
            self.append(key, value)
            
    def find(self, key):
        keyHash = hash(key) % self._N
        if not keyHash in self._hashes:
            raise Exception('Item not found')
        index = self._hashes[keyHash]
        return self._recursiveFind(key, index)

    def _recursiveFind(self, key, index):
        if self._values[index][0] == key:
            return self._values[index][1]
        
        if len(self._values[index]) < 3:
            raise Exception('Item not found')
            
        return self._recursiveFind(key, self._values[index][2])
            
    
    def printHashes(self):
        print("index => [key, value, pervious]:")
        for key, value in self._values.items():
            print("{} => {}".format(key,value))
            
        print("\nhash => index:")
        for key, value in self._hashes.items():
            print("{} => {}".format(key,value))
        
    def _checkCollisions(self, keyHash):
        return keyHash in self._hashes
    
    
try:
    arr = {"key1" : "value1", "key2" : "value2", "kkkwer2" : "vakue24"}
    chains = ChainsMethod(arr)
    chains.printHashes()
    
    print("\n\nappend key 'newKey'")
    chains.append("newKey", "newValue")
    chains.printHashes()
    
    print("\n\nfind by key 'key2'")
    print("result: ",chains.find("key2"))
    
    # тут будет ошибка, т.к ключа нет
    print("\n\nfind by key 'key'")
    print("result: ",chains.find("key"))
    
except Exception as inst:
    print(inst.args)
```
